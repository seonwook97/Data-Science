## 육아휴직관련 법안 대한민국 국회 제 1809890호 의안

```python
import nltk
```

```python
from konlpy.corpus import kobill

files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

- KoNLPy는 대한민국 법령을 가지고 있다

```python
doc_ko
```

```python
from konlpy.tag import Okt

t = Okt()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

- Twitter 엔진으로 명사 분석
- Twitter => Okt

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
```

```python
ko.tokens
```

```python
print(len(ko.tokens)) # returns number of tokens (document length)
print(len(set(ko.tokens))) # returns number of unique tokens
ko.vocab() # returns frequency distribution
```

- nltk를 사용해서 토큰(빈도수 포함) 분석

```python
import matplotlib.pyplot as plt
import set_matplotlib_hangul
```

```python
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', 'x', '의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를']
```

- 한글 stopword는 상황에 따라 복잡해서, 일단 손으로 그냥 잡음

```python
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')

plt.figure(figsize=(12, 6))
ko.plot(50) # Plot sorted frequency of top 50 tokens
plt.show()
```

```python
ko.count('초등학교')
```

- 특정 단어의 빈도수 조사

```python
plt.figure(figsize=(12, 6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

- 단어의 위치 확인

```python
ko.concordance('초등학교')
```

- 선택한 단어가 있는 문맥 파악

```python
ko.collocations()
```

- 영어에서 연어(collocation)란 무엇인가?
    - 연어란 '함께 위치하는 단어들'(co + location)이란 뜻으로 , 어휘의 조합 또는 짝을 이루는 말들을 일컫는다.    
    - 예를 들면 'friend'와 주로 같이 쓰이는 형용사는 'best', 'good', 그리고 'loyal'이다.
    - 'She talks to her best friend every day.'

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
data = ko.vocab().most_common(150)

wordcloud = WordCloud(
    font_path = 'C:/Windows/Fonts/malgun.ttf',
    relative_scaling=0.2, # 단어들 간의 간격
    background_color = 'white',    
).generate_from_frequencies(dict(data))

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
