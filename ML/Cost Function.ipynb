## Cost Function

### 실제 데이터(점 3개)와 구해야 할 모델(h)

- $ x_1 = (2,1), x_2 = (3,5), x_3 = (5,6) $ 
- $ h_\theta = \theta x $

#### 각각의 에러를 구해서 제곱


- 에러를 제곱하는 이유는 부호를 없애기 위함
- 그렇기 때문에 절대값을 사용하는 경우도 있음
- $(h_\theta(2) - 1)^2$, $(h_\theta(3) - 5)^2$, $(h_\theta(5) - 6)^2$ 

#### 이들의 평균을 구함
-  $\frac{1}{3}((h_\theta(2) - 1)^2 + (h_\theta(3) - 5)^2 + (h_\theta(5) - 6)^2)$ 

#### Cost Function
-  $J(\theta) = \frac{1}{3}((h_\theta(2) - 1)^2 + (h_\theta(3) - 5)^2 + (h_\theta(5) - 6)^2)$ 

#### Cost Fnc을 최소화할 수 있다면 최적의 직선을 찾을 수 있음
-  $J(\theta) = \frac{1}{3}((h_\theta(2) - 1)^2 + (h_\theta(3) - 5)^2 + (h_\theta(5) - 6)^2)$ 
- $minJ(\theta) --> h_\theta(x) = \theta x$

#### 계산
- $ h_\theta = \theta x $ <- 가정
- $J(\theta) = \frac{1}{3}((h_\theta(2) - 1)^2 + (h_\theta(3) - 5)^2 + (h_\theta(5) - 6)^2)$ 
- $J(\theta) = \frac{1}{3}(2\theta - 1)^2 + (3\theta - 5)^2 + (5\theta - 6)^2)$

$ J(\theta) = 38\theta^2 - 94\theta + 62 $ 

### 최솟값 찾기

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 6, 100)
y = 38*x**2 - 94*x + 62
y_diff = 1.24
```

```python
import matplotlib as mpl

mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('$x$',size=25)
plt.ylabel('$J(x) = 38x^2 - 94x + 62$', size=25)
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)
plt.show()
```
