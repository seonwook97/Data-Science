## XOR

- 선형 모델로는 XOR를 풀 수가 없다

```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1]
])

y = np.array([[0], [1], [1], [0]])
```

```python
X.shape, y.shape
```

### 모델은 간단히

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, activation='sigmoid', input_shape=(2, )),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

### model.compile

```python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1), loss='mse')
```

- 옵티마이저를 선정하고, 학습률을 선정한다
- loss 함수는 mse로 한다 mean squared error

### 학습

```python
hist = model.fit(X, y, epochs=5000, batch_size=1)
```

- epochs는 지정된 횟수만큼 학습을 하는 것
- batch_size는 한번의 학습에 사용될 데이터의 수를 지정

### 학습결과

```python
model.predict(X)
```

### loss 상황

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

### 학습에서 찾은 가중치

```python
for w in model.weights:
    print('---')
    print(w)
```

## iris 데이터

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
```

```python
y
```

### One-hot encoding

#### sklearn의 one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
enc.fit(y.reshape(len(y), 1))
```

```python
enc.categories_
```

```python
y_onehot = enc.transform(y.reshape(len(y), 1))
y_onehot[:3]
```

### 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot,
                                                   test_size=0.2,
                                                   random_state=13)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, input_shape=(4,), activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax'),
])
```

### activation
- 하나의 뉴런의 끝단에 activation이라는 아이가 붙어있음

### 역전파 back-propagation
- 뉴럴넷의 학습방법 Back propagation - 뒤로 전달
- 무엇을 전달하는가? - 현재 내가 틀린정도를 '미분(기울기)' 한 것
- 미분하고, 곱하고, 더하고를 역방향으로 반복하며 업데이트한다.

### 역전파에서는 sigmoid가 문제가 있다
- 우리가 activation 함수로 sigmoid를 썼다는 것
- 기울기가 0인 것들을 반복하여 중간에 곱하면 매우 작은 값이 됨 - gradient vanishing
- 레이어가 깊을수록 업데이트가 사라져간다. 그래서 fitting이 잘 안됨(underfitting)
- ReLU의 등장
- 사그라드는 sigmoid대신 죽지않는 activation func을 쓰자 -> ReLU

### softmax
- 뉴럴넷에게 답을 회신받는 3가지 방법
- Value -> output을 그냥 받는다.
- O/X -> output에 sigmoid를 먹인다.
- Category -> oupput에 softmax를 먹인다.

### adam

```python
model.compile(optimizer='adam', loss='categorical_crossentropy',
             metrics=['accuracy'])
model.summary()
```

#### gradient decent
- 기존 뉴럴넷이 가중치 parameter들을 최적화(optimize)하는 방법
- loss function의 현 가중치에서의 기울기(gradient)를 구해서 loss를 줄이는 방향으로 업데이트해 나간다.

#### 복습
- 뉴럴넷은 loss(or cost)function을 가지고 있다. 쉽게 말하면 '틀린 정도'
- 현재 가진 weight세팅(내 자리)에서, 내가 가진 데이터를 다 넣으면 전체 에러가 계산된다.
- 거기서 미분하면 에러를 줄이는 방향을 알 수 있다. (내자리의 기울기 * 반대방향)
- 그 방향으로 정해진 스텝량(learning rate)을 곱해서 weight을 이동시킨다. 이걸 반복
- weight의 업데이트 = 에러를 낮추는 방향(decent) * 한발자국 크기(learning rate) * 현 지점의 기울기(gradient)

#### SGD
- Gradient Decent는 데이터를 전부다 읽고나서 최적의 1스텝을 간다.
- SGD는 작은 토막마다 일단 1스텝을 간다.

#### GD vs SGD
- Gradient Decent
    - 모든걸 계산(1시간)후 최적의 한스텝 6스텝 * 1시간 = 6시간
    - 최적인데 너무 느리다

- Stochastic Gradient Descent
    - 일부만 검토(5분)
    - 틀려도 일단 간다 -> 빠른 스텝
    - 11스텝 * 5분 = 55분 < 1시간
    - 조금 헤매도 어쨌든 인근에 아주 빨리 갔다

#### 옵티마이저의 선택
- 산을 잘 타고 내려오는 것은 어느 **방향**으로 발을 디딜지 얼마 **보폭**으로 발을 디딜지
- 두가지를 잘 잡아야 빠르게 타고 내려온다.
- SGD를 더 개선한 멋진 optimizer가 많다. SGD의 개선된 후계자들
